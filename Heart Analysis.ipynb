## Cleaning Data

```python
# replacing all null values of the column Cal with the mean value
moy_Ca=heart["Ca"].mean()
heart["Ca"].fillna(moy_Ca,inplace=True)
```

```python
# replacing all null values of the column Thal with the fr√©quent value
freq_Thal=heart["Thal"].mode()[0]
heart.Thal.fillna(freq_Thal,inplace=True)
```

```python
# Here we make sure we fill all the null values
heart.info()
```

```python
# we have the first column with the label 0 that is'nt pertinent, we will drop it.
heart.drop(["Unnamed: 0"],axis=1, inplace=True)
```

```python
heart.head()
```

```python
# Now, we will make sure that we don't have duplicated values
heart.drop_duplicates(inplace=True)
```
